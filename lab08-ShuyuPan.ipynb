# Feature Engineering

Feature Engineering is a crucial step in preparing your data for machine learning. In this lab, we will focus on the following tasks:

1. **Drop rows with missing values** in the target variable and key features.
2. By now you are already familiar with the code and the data. Based on your understanding please choose any 3 (my code output has 10) variables as:
   1. two continuous variables (use your best judgment!)
   2. one categorical.
   3. Your dependent variable (y) is `SALARY`.
3. **Convert categorical variables** into numerical representations using **StringIndexer** and **OneHotEncoder**.
4. **Assemble features** into a single vector using **VectorAssembler**.
5. **Split the data** into training and testing sets.

````python
#| eval: false
#| echo: true
from pyspark.sql.functions import col
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml import Pipeline

df = df.dropna(subset=['SALARY', 'MIN_YEARS_EXPERIENCE', 'MAX_YEARS_EXPERIENCE', 'EMPLOYMENT_TYPE_NAME'])

categorical_cols = ['EMPLOYMENT_TYPE_NAME']

indexers = [StringIndexer(inputCol=col, outputCol=f"{col}_idx", handleInvalid='skip') for col in categorical_cols]
encoders = [OneHotEncoder(inputCol=f"{col}_idx", outputCol=f"{col}_vec") for col in categorical_cols]

assembler = VectorAssembler(
    inputCols=['MIN_YEARS_EXPERIENCE', 'MAX_YEARS_EXPERIENCE'] + [f"{col}_vec" for col in categorical_cols],
    outputCol="features"
)

pipeline = Pipeline(stages=indexers + encoders + [assembler])
data = pipeline.fit(df).transform(df)
data.show(5,False)
````

This section constructs a machine learning pipeline that processes categorical and numerical features using StringIndexer, OneHotEncoder, and VectorAssembler. The transformed DataFrame includes a features column that consolidates encoded categorical and scaled numerical inputs, preparing the data for model training.

# Diagnostic Plot

Diagnostic plots are essential for evaluating the performance of regression models. In this section, we will create several diagnostic plots to assess the linear regression model's assumptions and performance. There are four (2*2 grid) main plots we will create, you can use `seaborn` or `matplotlib` for this:

1. **Predicted vs Actual Plot**
2. **Residuals vs Predicted Plot**
3. **Histogram of Residuals**
4. **QQ Plot of Residuals**


````python
#| eval: true
#| echo: false
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Load predictions from GLR model
df_pred = summary.predictions.select("prediction", "SALARY")
df_pred = df_pred.toPandas()

# Compute residuals
df_pred["residuals"] = df_pred["SALARY"] - df_pred["prediction"]
df_pred["fitted"] = df_pred["prediction"]

# Standardized residuals
res_mean = df_pred["residuals"].mean()
res_std = df_pred["residuals"].std()
df_pred["std_residuals"] = (df_pred["residuals"] - res_mean) / res_std

# Square root of standardized residuals (for Scale-Location)
df_pred["sqrt_std_resid"] = np.sqrt(np.abs(df_pred["std_residuals"]))


plt.figure(figsize=(16, 12))
sns.set_theme(style="whitegrid")

# Plot 1: Predicted vs Actual Plot
plt.subplot(2, 2, 1)
sns.scatterplot(x="SALARY", y="fitted", data=df_pred)
plt.plot(
    [df_pred["SALARY"].min(), df_pred["SALARY"].max()],
    [df_pred["SALARY"].min(), df_pred["SALARY"].max()],
    'r--', label="Perfect Prediction"
)
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Predicted vs Actual Salary")
plt.legend()
plt.grid(True)

# Plot 2: Residuals vs Predicted Plot
plt.subplot(2, 2, 2)
sns.scatterplot(x="fitted", y="residuals", data=df_pred)
plt.axhline(0, color="red", linestyle="--", label="Zero Residual")
plt.xlabel("Predicted Salary (Fitted Values)")
plt.ylabel("Residuals")
plt.title("Residuals vs Predicted")
plt.legend()
plt.grid(True)

# Plot 3: Histogram of Residuals
plt.subplot(2, 2, 3)
sns.histplot(df_pred["residuals"], kde=True, bins=30, color="skyblue")
plt.axvline(0, color="red", linestyle="--", label="Zero Residual")
plt.xlabel("Residuals")
plt.ylabel("Frequency")
plt.title("Histogram of Residuals")
plt.legend()
plt.grid(True)

# Plot 4: QQ Plot of Residuals
plt.subplot(2, 2, 4)
stats.probplot(df_pred["residuals"], dist="norm", plot=plt)
plt.title("Normal Q-Q Plot of Residuals")
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Ordered Residuals")
plt.grid(True)

# # Plot 1: Residuals vs Fitted
# plt.subplot(2, 2, 1)

# Plot 2: Normal Q-Q
# plt.subplot(2, 2, 2)

# Plot 3: Scale-Location
# plt.subplot(2, 2, 3)

# Plot 4: Residuals vs Leverage â Approximate
# Note: Leverage & Cook's Distance require X matrix; we approximate using fitted & residual
# plt.subplot(2, 2, 4)

plt.tight_layout()
plt.savefig("_output/glr_diagnostic_classic.png")
plt.show()
````

Plot 1: This plot compares the GLR modelâs predicted salaries against actual values. The wide scatter below the ideal fit line suggests the model underpredicts salaries for many observations and lacks high predictive accuracy.

Plot 2: The residuals appear randomly scattered around the zero line with increasing spread, indicating possible heteroscedasticity. This violates a key assumption of linear regression, suggesting the modelâs variance is not constant across predictions.

Plot 3: The distribution of residuals is approximately normal but slightly left-skewed, with a long tail extending into negative residuals. This skew may indicate bias in the model's errors, especially for lower-salary predictions.

Plot 4: The Q-Q plot shows that most residuals follow a normal distribution, but deviations at the tails indicate potential outliers or heavy tails. This could signal that extreme salary values are not well-modeled by the GLR approach.

# Evaluation

The evaluation of the model is crucial to understand its performance. In this section, we will calculate and visualize the following metrics:
1. **RÂ² (Coefficient of Determination)**: Indicates how well the model explains the variance in the target variable.
2. **RMSE (Root Mean Squared Error)**: Measures the average magnitude of the errors between predicted and actual values.

````python
#| eval: true
#| echo: false
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.sql.functions import col, pow, sqrt, avg
import numpy as np
from pyspark.ml.regression import GeneralizedLinearRegression

# Train GLR model
glr = GeneralizedLinearRegression(featuresCol='features', labelCol='SALARY', family='gaussian', link='identity')
glr_model = glr.fit(train_data)
summary = glr_model.summary

pred_glr = glr_model.transform(test_data)

# RÂ²
r2_eval = RegressionEvaluator(labelCol="SALARY", predictionCol="prediction", metricName="r2")
r2 = r2_eval.evaluate(pred_glr)
# AIC from GLR summary
aic = summary.aic

# BIC calculation
n = test_data.count()
k = len(glr_model.coefficients)
rss = summary.deviance 
bic = n * np.log(rss / n) + k * np.log(n)

# RMSE manually
residuals_df = pred_glr.withColumn("squared_error", pow(col("SALARY") - col("prediction"), 2))
rmse = np.sqrt(residuals_df.select(avg("squared_error")).first()[0])
````

The model evaluation computed key performance metrics including RÂ², RMSE, AIC, and BIC to assess fit quality and prediction error. A relatively low RÂ² combined with a high RMSE suggests that the model explains only a small portion of the variance in salary, indicating potential room for improvement through feature selection or model tuning.
